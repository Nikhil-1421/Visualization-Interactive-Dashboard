```python
# From viewing dr_df we can see that Lewis Hamilton's unique driverID is 1 - so now I can create a new dataframe of Lewis's qualifying performances from qli_df

Lewis_df = qli_df[qli_df['driverId'] == 1].copy()
Lewis_df
```

```python
# The only columns I need are the raceID, q1, q2, q3 and position columns, so I filtered out the rest 

columns_to_drop = ['qualifyId','driverId','constructorId','number']
Lewis_df.drop(columns=columns_to_drop,inplace=True)
Lewis_df
```

```python
# In order to match the raceID with the Grand Prix name, I need to filter out rc_df, so I created a new dataframe that I will drop the unneeded columns from.

race_df = rc_df

col1 = ['round','circuitId','date','time','url','fp1_date','fp1_time','fp2_date','fp2_time','fp3_date','fp3_time','quali_date','quali_time','sprint_date','sprint_time']
race_df.drop(columns=col1,inplace=True)
race_df
```

```python
# Now I can take the two filtered dataframes and create one new one that will display the information I need

merged_df = Lewis_df.merge(race_df[['raceId', 'year', 'name']], how='left', on='raceId')
merged_df.drop('raceId', axis=1, inplace=True)
merged_df.rename(columns={'year': 'year', 'name': 'Grand Prix'}, inplace=True)
merged_df
```

```python
# Exporting our new dataframe to csv format

merged_df.to_csv('/kaggle/working/output.csv', index=False)
```
